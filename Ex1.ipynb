# One variable linear regression

### Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=np.loadtxt('ex1data1.txt', delimiter=',')
```

```python
X = np.c_[np.ones(data.shape[0]),data[:,0]]
Y = np.c_[data[:,1]]

iterations=15000
alpha=0.001
theta=np.array([[0.0],[0.0]])

plt.figure(1)
plt.plot(X[:,1],Y,'ro')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')

```

### Computing the cost

```python
def h(X):
    return np.dot(X,theta)
```

```python
def j(X):
    m=X.shape[0]
    return 1/(2*m)*sum((h(X)-Y)**2)
```

```python
j(X) # expected 32.07
```

### Grandient descent

```python
def gd(X):
    m=X.shape[0]
    tJ=[0,0]
    for j in range(2):
        sumH=0
        for i in range(X.shape[0]):
            sumH+=(h(X[i])-Y[i])*X[i][j]
        tJ[j]=theta[j]-alpha/m*sumH
        
    theta[0]=tJ[0]
    theta[1]=tJ[1]
    
```

```python
cost=np.zeros(iterations)

for i in range(iterations):
    gd(X)
    cost[i]=j(X)

plt.figure(2)
plt.plot(X,h(X))
plt.plot(X[:,1],Y,'ro')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')

plt.figure(3)
plt.plot(cost)
plt.xlabel('# of iterations')
plt.ylabel('j(0)')
```
